```python
import matplotlib.pyplot as pd

KursUSD = 4
prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]
pricesUSD = []
for price in prices:
    pricesUSD.append(price[1] / KursUSD)    #Вираховуємо ціни в USD по заданому курсу
pd.grid(visible=None, which='major', axis='both')
pd.plot(pricesUSD, linestyle='dashed')
```
