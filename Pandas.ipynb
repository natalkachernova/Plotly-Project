```python
import pandas as pd

KursUSD = 4
prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]
pricesUSD = []
df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
for price in prices:
    pricesUSD.append(price[1] / KursUSD)    #Вираховуємо ціни в USD по заданому курсу
df['pricesUSD'] = pricesUSD                 #Додаємо третю колонку з цінами в USD
df['pricesUSD'].plot(kind='line', linestyle='dashed', linewidth=2, markersize=12, grid='true')
df
```
